```python
import pandas as pd
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
df = pd.read_csv('imports-85.csv')
```

```python
df.head()
```

```python
df_novo = df[['make','wheel-base', 'curb-weight', 'engine-size', 
              'horsepower', 'city-mpg', 'highway-mpg']]
```

```python
df_novo.head()
```

```python
sns.pairplot(df_novo, hue='make')
```

```python
x = np.array(df['curb-weight']).reshape((-1, 1))
y = np.array(df['city-mpg'])
```

```python
x, y
```

```python
sns.lmplot(x='curb-weight',y='city-mpg', data=df_novo)
```

```python
#Gerar modelo de regressão
mdl = LinearRegression().fit(x,y) 
```

```python
#Calcular R-Sq
r_sq = mdl.score(x,y)
```

```python
print('o coeficiente R_SQ é {0:.2f}%'.format(r_sq*100))
```

```python
#Calcular o coeficiente (angulação da reta) e a constante (intercept) 
coef = mdl.coef_
const = mdl.intercept_

print ("city-mpg = {0:.3f} + {1:.3f} curb-weight".format(const,coef[0]))
```
